```python
from faker import Faker
import random
import pandas as pd

fake = Faker('en_IN')  # Setting locale to Indian English for more relevant data

# Function to generate random budget based on base_budget and number of parameters
def generate_budget(base_budget, num_params):
    return base_budget + (num_params * 100)

# Generate 500 rows of synthetic data
data = []
for _ in range(3000):
    student_id = fake.random_number(digits=3)
    name = fake.name()
    gender = random.choice(['Male', 'Female'])
    university = fake.company()
    room_type = random.choice(['Single', 'Double', 'Copper'])
    base_budget = 2000
    num_params = random.randint(0, 5)  # Number of additional parameters
    budget = generate_budget(base_budget, num_params)
    district = random.choice([ "Alipurduar", "Bankura", "Birbhum", "Cooch Behar", "Dakshin Dinajpur",
    "Darjeeling", "Hooghly", "Howrah", "Jalpaiguri", "Jhargram", "Kalimpong",
    "Kolkata", "Malda", "Murshidabad", "Nadia", "North 24 Parganas",
    "Paschim Medinipur", "Paschim Bardhaman", "Purba Bardhaman",
    "Purba Medinipur", "Purulia", "South 24 Parganas", "Uttar Dinajpur"])
    location = fake.city()
    area = fake.street_name()
    pin = fake.postcode()  # Using postcode() instead of zipcode()
    near_rail_station = random.choice(['Yes', 'No'])
    near_university = random.choice(['Yes', 'No'])
    near_school = random.choice(['Yes', 'No'])
    near_roadside = random.choice(['Yes', 'No'])
    pen_guest = random.choice(['Yes', 'No'])
    mess = 'Yes' if pen_guest == 'No' else 'No'
    room_info = random.choice(['Yes', 'No'])  # Information about room availability
    
    data.append([student_id, name, gender, university, room_type, budget, district, location, area, pin, near_rail_station, near_university, near_school, near_roadside, mess, pen_guest, room_info])

# Create DataFrame
df = pd.DataFrame(data, columns=['Student ID', 'Name', 'Gender', 'University', 'Room Type', 'Budget', 'District', 'Location', 'Area', 'Pin', 'Near Rail Station', 'Near University', 'Near School', 'Near Roadside', 'Availability of Mess', 'Availability of Pen Guest', 'Room Information'])

# Save to CSV
df.to_csv('synthetic_data_west_bengal_updated.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('synthetic_data_west_bengal_updated.csv')

# Set style for plots
sns.set_style("whitegrid")
```

```python

# Plot 1: Count of Rooms by Room Type
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Room Type')
plt.title('Count of Rooms by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot 2: Count of Rooms by District
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='District', order=df['District'].value_counts().index)
plt.title('Count of Rooms by District')
plt.xlabel('District')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot 3: Count of Rooms by Availability of Room Information
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Room Information')
plt.title('Count of Rooms by Availability of Room Information')
plt.xlabel('Availability of Room Information')
plt.ylabel('Count')
plt.show()
```

```python
# Plot 4: Count of Rooms by Room Type and Availability of Room Information
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Room Type', hue='Room Information')
plt.title('Count of Rooms by Room Type and Availability of Room Information')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.legend(title='Room Information')
plt.xticks(rotation=45)
plt.show()
```
